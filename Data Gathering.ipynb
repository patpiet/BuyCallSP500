```python
import pandas as pd
import yfinance as yf
```

```python
# Get max range data of SPX500
df = yf.download("SPY")
```

```python
df
```

```python
# Library for the technical indicators
import ta
```

```python
# Simple Moving Average Indicator (SMA)
df['SMA'] = ta.trend.SMAIndicator(df['Close'], window=14).sma_indicator()
```

```python
# Moving Average Convergence Divergence (MACD)
df['MACD'] = ta.trend.MACD(df['Close'], window_sign=14).macd()
```

```python
# Relative Strength Index (RSI)
df['RSI'] = ta.momentum.RSIIndicator(df['Close'], window=14).rsi()
```

```python
# Percentage Price Oscillator indicator (PPO)
ta.momentum.ppo(df['Close'], window_sign=14)
```

```python
# Average Directional Index (ADX)
df['ADX'] = ta.trend.ADXIndicator(df['High'], df['Low'], df['Close'], window=14).adx()
```

```python
# Williams Percent Range (%R)
df['R%'] = ta.momentum.WilliamsRIndicator(df['High'], df['Low'], df['Close'], lbp=14).williams_r()
```

```python
# Accumulation/Distribution Line Indicator (A/D)
df['ADI'] = ta.volume.acc_dist_index(df['High'], df['Low'], df['Close'], df['Volume'])
```

```python
#  Aroon Oscillator (AO)
df['AI'] = ta.trend.AroonIndicator(df['Close'], window=14).aroon_indicator()
```

```python
# Commodity Channel Index (CCI)
df['CCI'] = ta.trend.CCIIndicator(df['High'], df['Low'], df['Close'], window=14).cci()
```

```python
df.to_csv('SPYIndicators.csv')
```
