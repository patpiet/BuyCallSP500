```python
df = pd.read_csv('SPYIndicators.csv')
```

```python
# The only missing values are the first few days/weeks as some of the indicators needed prior knowledge
sns.heatmap(df.isna().transpose(),
            cmap="YlGnBu",
            cbar_kws={'label': 'Missing Data'})
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
import datetime

def isWorth(df, weeks=4):
    for index, row in df.iterrows():
        price = row['Close']
        date_plus_days = df['Date'] + datetime.timedelta(weeks=weeks)
        price_plus_days = df[df['Date'] == date_plus_days]['Close']
        print(price)
        print(price_plus_days)
        print('-----')
        
```
